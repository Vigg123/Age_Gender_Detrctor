```python
 

# Read the image using OpenCV
img1 = cv2.imread(r"C:\Users\vigne\OneDrive\Desktop\Age gender deccteter\data\images.jpeg")

# Check if the image was loaded properly
if img1 is None:
    raise ValueError("Image not found. Please check the path.")

# Convert the image from BGR to HSV
img_hsv = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)

# Apply blur to the HSV image
figure_size = 7
result = cv2.blur(img_hsv, (figure_size, figure_size))

# Convert back to RGB for display purposes
img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
result_rgb = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)

# Display the images using Matplotlib
plt.figure(figsize=(11, 6))

plt.subplot(121)
plt.imshow(img_rgb)
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122)  # Changed from 131 to 122 for proper subplot placement
plt.imshow(result_rgb)
plt.title('Blurred')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img1 = cv2.imread(r"C:\Users\vigne\OneDrive\Desktop\Age gender deccteter\data\images.jpeg")

# Get the dimensions of the image
(height, width) = img1.shape[0:2]

# Resize the image
zoom = cv2.resize(img1, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)

# Convert BGR image to RGB
zoom_rgb = cv2.cvtColor(zoom, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(zoom_rgb)
plt.axis('off')  # Hide axis
plt.show()
```

```python
zoom = cv2.resize(img1,(int(width*2),int(height*2)), interpolation=cv2.INTER_CUBIC)
zoom_rgb = cv2.cvtColor(zoom, cv2.COLOR_BGR2RGB)
plt.imshow(zoom_rgb)
```
